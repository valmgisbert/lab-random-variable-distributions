```python
# import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../lab-cleaning-numerical-data/numerical.csv')
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data
```

```python
# plot cols
def plot_df(df):
    for col in df.columns:
        sns.displot(df[col], kde=True)
        plt.title(col)
        plt.show()

plot_df(data)
```

**Comments:**
- There are no normal distributions in sight.
    - Potentially exponential: `total_claim_amount` and `income`; even with the spike of the first one, they show a constant decrease in the plot. We also see `number_of_open_comlaints` and `number_of_policies`, that even though it can qualify as slightly categorical we still see the linear decrease as it goes.


```python
# use log transform on the previous cols to check for any changes
num_log = np.log(data[['total_claim_amount','income','number_of_open_complaints','number_of_policies']])

plot_df(num_log)
```
